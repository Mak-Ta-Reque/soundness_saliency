```python
import os
import glob
from PIL import Image
import numpy as np
```

```python
path = "/mnt/sda/abka03-data/imagenet/imagenette2-320/val/*/*.JPEG"
```

```python
files = glob.glob(path)
```

```python
image_net_file = "imagenet.txt"
imagenmnet = open(image_net_file)#.split('\n')
```

```python
data = {}

# Loop through each line of the text
for line in imagenmnet:
    # Split the line by whitespace to get the key-value pairs
    parts = line.split()
    # Add the key-value pair to the dictionary
    data[parts[1]] = [parts[0], ' '.join(parts[2:])]
# Print the resulting dictionary

np.save("imagenet_label_name.npy", data)
```

```python
images = []
labels = []
for item in files:
    image = Image.open(item)
    width, height = image.size
    new_size = min(width, height)
    left = (width - new_size) // 2
    top = (height - new_size) // 2
    right = (width + new_size) // 2
    bottom = (height + new_size) // 2
    img = image.crop((left, top, right, bottom))

    # Resize the image to (224, 224)
    img = img.resize((224, 224), resample=Image.NEAREST)

    # Transpose the image
    img = np.array(img)
    
```

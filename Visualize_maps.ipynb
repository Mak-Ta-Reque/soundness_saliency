```python
import os
import numpy as np
from matplotlib import pyplot as plt
```

```python
os.listdir('results')
```

```python
images = np.load('input_data/sample_images.npy', allow_pickle=True)
labels = np.load('input_data/sample_labels.npy', allow_pickle=True)
```

```python
label_list = [374, 520, 534, 140, 667, 282, 532, 282, 0, 790, 161, 990]
maps = {}
for label in label_list:
    maps[label] =np.squeeze(np.load('results/batch_0label_' +str(label) + '.pkl', allow_pickle=True)[('rand_image', 1, 4, 2e-05, 0.01, 0.0)]) 
```

```python
fig = plt.figure(figsize = (10,10))
for k in range(1,11,2):
    plt.subplot(5,2,k)
    lbl = labels[k-1]
    plt.imshow(maps[lbl][k-1])
    plt.subplot(5,2,k+1)
    plt.imshow(np.transpose(images[k-1], [1,2,0]))
```
